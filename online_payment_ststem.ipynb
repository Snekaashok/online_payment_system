```python
import pandas as pd

# Sample data (replace with actual data)
payment_systems_data = {
    'PaymentSystem': ['PayPal', 'Stripe', 'Square'],
    'TransactionFees': [0.03, 0.029, 0.025],  # Transaction fees in decimal format (e.g., 3% = 0.03)
    'ProcessingTime (seconds)': [2.5, 1.8, 2.2]  # Average transaction processing time in seconds
}

# Create DataFrame
df = pd.DataFrame(payment_systems_data)

# Display the data
print("Payment Systems Comparative Data:")
print(df)
```

```python
import matplotlib.pyplot as plt

# Plot transaction fees
plt.figure(figsize=(10, 5))
plt.bar(df['PaymentSystem'], df['TransactionFees'], color=['blue', 'green', 'orange'])
plt.xlabel('Payment System')
plt.ylabel('Transaction Fees')
plt.title('Transaction Fees Comparison')
plt.show()

# Plot processing time
plt.figure(figsize=(10, 5))
plt.bar(df['PaymentSystem'], df['ProcessingTime (seconds)'], color=['blue', 'green', 'orange'])
plt.xlabel('Payment System')
plt.ylabel('Processing Time (seconds)')
plt.title('Transaction Processing Time Comparison')
plt.show()
```
